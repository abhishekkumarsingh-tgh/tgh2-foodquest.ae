```sql
%python
catalogs = [row.catalog for row in spark.sql("SHOW CATALOGS").collect()]
print(catalogs)
```

```sql
%python
catalog = 'fq_dev_catalog'
print(f"Catalog: {catalog}")
schemas = [row.databaseName for row in spark.sql(f"SHOW SCHEMAS IN {catalog}").collect()]
for schema in schemas:
    print(f"  Schema: {schema}")
    tables = [row.tableName for row in spark.sql(f"SHOW TABLES IN {catalog}.{schema}").collect()]
    for table in tables:
        print(f"    Table: {table}")
```

```sql
%python
users = ['user1@foodquest.ae', 'user2@foodquest.ae', 'user3@foodquest.ae']

for user in users:
    spark.sql(f'GRANT USAGE ON SCHEMA fq_dev_catalog.silver TO "{user}"')
    spark.sql(f'GRANT SELECT, INSERT, UPDATE, DELETE ON ALL TABLES IN SCHEMA fq_dev_catalog.silver TO "{user}"')
    spark.sql(f'ALTER DEFAULT PRIVILEGES IN SCHEMA fq_dev_catalog.silver GRANT SELECT, INSERT, UPDATE, DELETE ON TABLES TO "{user}"')
    spark.sql(f'GRANT USAGE ON ALL SEQUENCES IN SCHEMA fq_dev_catalog.silver TO "{user}"')
```

```python
bronze_path = spark.sql(f"""DESCRIBE EXTERNAL LOCATION `{env}_extloc_bronze`""").select("url").collect()[0][0]
silver_path = spark.sql(f"""DESCRIBE EXTERNAL LOCATION `{env}_extloc_silver`""").select("url").collect()[0][0]
gold_path = spark.sql(f"""DESCRIBE EXTERNAL LOCATION `{env}_extloc_gold`""").select("url").collect()[0][0]
staging_path = spark.sql(f"""DESCRIBE EXTERNAL LOCATION `{env}_extloc_staging`""").select("url").collect()[0][0]
catalog_path = spark.sql(f"""DESCRIBE EXTERNAL LOCATION `{env}_extloc_catalog`""").select("url").collect()[0][0]
```

```python
print(f'Bronze Path: {bronze_path}')
print(f'Silver Path: {silver_path}')
print(f'Gold Path: {gold_path}')
print(f'Staging Paht: {staging_path}')
print(f'Catalog Path: {catalog_path}')
```

```python
def create_dev_catalog(environment, catalog_path):
    print(f'Creating {environment}_catalog with managed storage')
    spark.sql(
        f"""
        CREATE CATALOG IF NOT EXISTS {environment}_catalog
        MANAGED LOCATION '{catalog_path}'
        """
    )
```

```python
create_dev_catalog(env, catalog_path)
```

```python
def create_Bronze_Schema(env,bronze_path):
    print(f'Using {env}_Catalog ')
    spark.sql(f""" USE CATALOG '{env}_catalog'""")
    print(f'Creating Bronze Schema in {env}_Catalog')
    spark.sql(f"""CREATE SCHEMA IF NOT EXISTS `bronze` MANAGED LOCATION '{bronze_path}'""")
```

```python
spark.sql(f""" USE CATALOG '{env}_catalog'""")
spark.sql(f"""CREATE SCHEMA IF NOT EXISTS `bronze` MANAGED LOCATION '{bronze_path}'""")
```

```python
create_Bronze_Schema(env,bronze_path)
```

```python
def create_Silver_Schema(environment,path):
    print(f'Using {environment}_Catalog ')
    spark.sql(f""" USE CATALOG '{environment}_catalog'""")
    print(f'Creating Silver Schema in {environment}_Catalog')
    spark.sql(f"""CREATE SCHEMA IF NOT EXISTS `silver` MANAGED LOCATION '{path}'""")
```

```python
create_Silver_Schema(env,silver_path)
```

```python
def create_Gold_Schema(environment,path):
    print(f'Using {environment}_Catalog ')
    spark.sql(f""" USE CATALOG '{environment}_catalog'""")
    print(f'Creating Gold Schema in {environment}_Catalog')
    spark.sql(f"""CREATE SCHEMA IF NOT EXISTS `gold` MANAGED LOCATION '{path}'""")
    print(f'****Schema creation successfull ! *****')
```

```python
create_Gold_Schema(env,gold_path)
```
